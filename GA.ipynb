```python
from MicropolisControlScript import Quimby
import numpy as np
from gi.repository import Gtk as gtk
import json
import random
```

# Genetic Algorithm to build micropolis city

```python
q = Quimby(map_w=30,map_h=30,gen_size=100,n_population=100,n_steps_evaluation=10)
```

```python
%%capture
p_selection = 0.2
n_mutations = 10
best_pop = []
for i in range(10) :
    q.evaluate_all()
    ranked_genomes = [x for _,x in sorted(zip(q.populations,q.genomes))]
    parents = ranked_genomes[:int(round(len(ranked_genomes)*p_selection))]
    children = list()
    best_pop.append(np.max(q.populations))
    while len(children) < q.n_population :
        i = np.random.randint(0,len(parents))
        j = np.random.randint(0,len(parents))
        if i!=j : 
            children.append(q.mate(parents[i],parents[j]))
    # Apply mutations
    idxs = list(range(q.n_population))
    random.shuffle(idxs)
    for a in idxs[0:n_mutations] : 
        q.mutate_genome(children[a])
    q.genomes = children
```
